Add more Dense layers to the existing code and check how the accuracy changes.

```python
import keras
import pandas
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation

# load dataset
from sklearn.model_selection import train_test_split
import pandas as pd
import numpy as np

dataset = pd.read_csv(path_to_csv, header=None).values

X_train, X_test, Y_train, Y_test = train_test_split(dataset[:,0:8], dataset[:,8],
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
my_first_nn = Sequential() # create model
my_first_nn.add(Dense(20, input_dim=8, activation='relu')) # hidden layer1
my_first_nn.add(Dense(15,activation='relu'))
my_first_nn.add(Dense(8,activation='relu'))
my_first_nn.add(Dense(1, activation='sigmoid')) # output layer
my_first_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['acc'])
my_first_nn_fitted = my_first_nn.fit(X_train, Y_train, epochs=100, initial_epoch=0)
print(my_first_nn.summary())
print(my_first_nn.evaluate(X_test, Y_test))
```

Change the data source to Breast Cancer dataset available in the source code folder and make required
changes. Report accuracy of the model.

```python
breast_cancer_csv = pd.read_csv('/content/gdrive/MyDrive/breastcancer.csv')

# Drop irrelevant columns
breast_cancer_csv.drop(['id', 'Unnamed: 32'], axis=1, inplace=True)

# Convert labels to binary
breast_cancer_csv['diagnosis'] = breast_cancer_csv['diagnosis'].map({'M': 1, 'B': 0})

X = breast_cancer_csv.drop('diagnosis', axis=1)
y = breast_cancer_csv['diagnosis']
X_train, X_test, Y_train, Y_test = train_test_split(X, y,
                                                    test_size=0.25, random_state=87)
np.random.seed(155)
breast_cancer_nn = Sequential() # create model
breast_cancer_nn.add(Dense(20, input_dim=X_train.shape[1], activation='relu')) # hidden layer1
breast_cancer_nn.add(Dense(15,activation='relu'))
breast_cancer_nn.add(Dense(8,activation='relu'))
breast_cancer_nn.add(Dense(1, activation='sigmoid')) # output layer
breast_cancer_nn.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
history = breast_cancer_nn.fit(X_train, Y_train, epochs=100, validation_split=0.2, initial_epoch=0)
print(breast_cancer_nn.summary())
print(breast_cancer_nn.evaluate(X_test, Y_test))
```
